```python
import tkinter as tk

def draw_points(canvas, xc, yc, x, y):
    for dx, dy in [(x, y), (-x, y), (x, -y), (-x, -y),
                   (y, x), (-y, x), (y, -x), (-y, -x)]:
        canvas.create_oval(xc + dx, yc + dy, xc + dx + 1, yc + dy + 1,
                           fill="blue", outline="blue")

def midpoint_circle(canvas, xc, yc, r):
    x = 0
    y = r
    p = 1 - r
    draw_points(canvas, xc, yc, x, y)

    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1
        draw_points(canvas, xc, yc, x, y)

def draw_circle():
    root = tk.Tk()
    root.title("Midpoint Circle Drawing - Blue Circle")

    canvas = tk.Canvas(root, width=600, height=600, bg="white")
    canvas.pack()

    xc = 300
    yc = 300
    r = int(input("Enter radius: "))

    canvas.create_text(300, 40, text=f"Center: ({xc},{yc}) | Radius: {r}",
                       font=("Arial", 12), fill="blue")

    midpoint_circle(canvas, xc, yc, r)

    canvas.create_oval(xc - 3, yc - 3, xc + 3, yc + 3, fill="blue")

    root.mainloop()

draw_circle()
```
